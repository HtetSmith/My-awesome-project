```python
final_merged_df['content'] = final_merged_df['genres'] + ' ' + final_merged_df['developer'] + ' ' + final_merged_df['steamspy_tags'] + ' ' + final_merged_df['short_description']
```

```python
# Transform the combined text data using TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(final_merged_df['content'])
```

```python
from sklearn.metrics.pairwise import linear_kernel

cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
# Function to get recommendations
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = final_merged_df[final_merged_df['name'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # Top 10 similar games
    game_indices = [i[0] for i in sim_scores]
    return final_merged_df.iloc[game_indices]
```

```python
# Example usage
recommendations = get_recommendations('Portal 2')
print(recommendations)
```
